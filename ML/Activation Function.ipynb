# Activation Function

Reference: [How to Choose an Activation Function for DL](https://machinelearningmastery.com/choose-an-activation-function-for-deep-learning/)

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# define input data
inputs = [x for x in range(-10, 10)]
```

```python
# plot inputs vs outputs
def plot(inputs, outputs):
    plt.plot(inputs, outputs)
    plt.show()
```

## Activation for Hidden Layers

### ReLU

The ReLU function is calculated as follows: $max(0.0, x)$.

```python
# rectified linear function
def rectified(x):
    return max(0.0, x)

# calculate outputs
outputs = [rectified(x) for x in inputs]

# plot inputs vs outputs
plot(inputs, outputs)
```

When using the ReLU function for hidden layers, it is a good practice to use a __“He Normal” or “He Uniform” weight initialization__ and __scale input data to the range 0-1 (normalize)__ prior to training.

### Sigmoid
= logistic function

The function takes any real value as input and outputs values in the range 0 to 1. The larger the input (more positive), the closer the output value will be to 1.0, whereas the smaller the input (more negative), the closer the output will be to 0.0.

The sigmoid activation function is calculated as follows: $1.0 / (1.0 + e^{-x})$

```python
# sigmoid activation function
def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))

# calculate outputs
outputs = [sigmoid(x) for x in inputs]

# plot inputs vs outputs
plot(inputs, outputs)
```

S-shape

When using the Sigmoid function for hidden layers, it is a good practice to use a __“Xavier Normal” or “Xavier Uniform” weight initialization__ (also referred to Glorot initialization, named for Xavier Glorot) and __scale input data to the range 0-1__ (e.g. the range of the activation function) prior to training.

### Tanh
= hyperbolic tangent = tanh = TanH

It is very similar to the sigmoid activation function and even has the same S-shape.

The function takes any real value as input and outputs values in the range -1 to 1. The larger the input (more positive), the closer the output value will be to 1.0, whereas the smaller the input (more negative), the closer the output will be to -1.0.

The Tanh activation function is calculated as follows:
$(e^{x} – e^{-x}) / (e^{x} + e^{-x})$

```python
# tanh activation function
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

# calculate outputs
outputs = [tanh(x) for x in inputs]

# plot inputs vs outputs
plot(inputs, outputs)
```

S-shape

When using the TanH function for hidden layers, it is a good practice to use a __“Xavier Normal” or “Xavier Uniform” weight initialization__ (also referred to Glorot initialization, named for Xavier Glorot) and __scale input data to the range -1 to 1__ (e.g. the range of the activation function) prior to training.

## Activation for Output Layers

### Linear
= identify (multiplied by 1.0) = no activation

The linear activation function does not change the weighted sum of the input in any way and instead returns the value directly.

```python
# linear activation function
def linear(x):
    return x

# calculate outputs
outputs = [linear(x) for x in inputs]

# plot inputs vs outputs
plot(inputs, outputs)
```

diagonal line shape

Target values used to train a model with a linear activation function in the output layer are typically scaled prior to modeling using __normalization or standardization transforms__.

### Sigmoid (same as above)

```python
# sigmoid activation function
def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))

# calculate outputs
outputs = [sigmoid(x) for x in inputs]

# plot inputs vs outputs
plot(inputs, outputs)
```

Target labels used to train a model with a sigmoid activation function in the output layer will have the values 0 or 1.

### Softmax
The softmax function outputs a vector of values that sum to 1.0 that can be interpreted as probabilities of class membership.

It is related to the argmax function that outputs a 0 for all options and 1 for the chosen option. Softmax is a “softer” version of argmax that allows a probability-like output of a winner-take-all function.

The softmax function is calculated as follows: $e^{x} / sum(e^{x})$

```python
# softmax activation function
def softmax(x):
    return np.exp(x) / np.exp(x).sum()
 
# define input data
softmax_inputs = [1.0, 3.0, 2.0]

# calculate outputs
outputs = softmax(softmax_inputs)

# report the probabilities
print(outputs)

# report the sum of the probabilities
print(outputs.sum())
```

Target labels used to train a model with the softmax activation function in the output layer will be vectors with 1 for the target class and 0 for all other classes.
